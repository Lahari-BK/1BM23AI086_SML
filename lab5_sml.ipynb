```python
import pandas as pd
from scipy.stats import chi2_contingency
data = {
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female"],
    "VotingPreference": ["A", "B", "A", "B", "A", "A", "B", "B"]
}
df = pd.DataFrame(data)
contingency_table = pd.crosstab(df["Gender"], df["VotingPreference"])

print("Contingency Table:")
print(contingency_table)

chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

print("\nChi-squared Statistic:", chi2_stat)
print("p-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)
if p_value < 0.05:
    print("\nConclusion: There is evidence to reject the null hypothesis.")
else:
    print("\nConclusion: There is insufficient evidence to reject the null hypothesis.")
```
